```python
def factorial(n):
    # Esta función tiene ERROR - range incorrecto
    if n == 0:
        return 1
    result = 1
    for i in range(1, n):  # ERROR: debería ser n+1
        result *= i
    return result
```

```python
# TEST EJERCICIO 2 - 10 PUNTOS
assert factorial(0) == 1
assert factorial(1) == 1
# Los siguientes tests fallarán debido al error
try:
    assert factorial(5) == 120
    print("Test 3: OK!")
except:
    print("Test 3: ERROR! - factorial(5) debería ser 120")
    
print("Ejercicio 2: 2.5/10 puntos")
print("Total estimado: 7.5/15 puntos")
```
